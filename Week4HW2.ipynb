```python
import pandas as pd
```

```python
years = pd.date_range(start="2022", end="2032", freq="Y")
years
```

```python
df = pd.DataFrame(columns=["revenue", "EBITDA", "taxes",
                           "cap_ex", "work_cap",
                           "cash_flow", "discount_factor",
                           "present_value_fcf"], index=years)
df
```

```python
df.loc["2022", "revenue"] = 222 * (1 + 0.18)
df
```

```python
for i in range(1, len(df)):
    df.iloc[i]["revenue"] = df.iloc[i - 1]["revenue"] * (1.18)
```

```python
df
```

```python
df["EBITDA"] = df["revenue"] * 0.45
df
```

```python
df["taxes"] = -df["EBITDA"] * 0.21
df
```

```python
df["cap_ex"] = -df["revenue"] * 0.03
df
```

```python
df["work_cap"] = -df["revenue"] * 0.06
df
```

```python
df["cash_flow"] = df["EBITDA"] + df["taxes"] + df["cap_ex"] + df["work_cap"]
df
```

```python
df.iloc[0]["discount_factor"] = 0.94
df
```

```python
# (1 + DF) ^ (-N)
for i in range(1, len(df)):
    df.iloc[i]["discount_factor"] = 1 / ((1 + 0.06) ** (i + 1))
df
```

```python
revenue_2032 = df.loc["2031", "revenue"].values[0] * (1.18)
revenue_2032
```

```python
EBITDA_2032 = revenue_2032 * 0.45
EBITDA_2032
```

```python
terminal_value = EBITDA_2032 * 32.70
terminal_value
```

```python
present_teminal_value = terminal_value * df.loc["2031", "discount_factor"].values[0]
present_teminal_value
```

```python
present_value_of_projected_fcf = df["cash_flow"].sum()
present_value_of_projected_fcf
```

```python
output = f"""
Our projected Present Value of Projected Future Cash Flows are: {present_value_of_projected_fcf}.

We are providing revenue guidance in 2032 to be: {revenue_2032}.

With an estimated EBITDA around: {EBITDA_2032}.

The terminal value a this time given a multiple of 32.70

and the above EBITDA for 2030, would equal: {present_teminal_value}."

"""
print(output)
```
